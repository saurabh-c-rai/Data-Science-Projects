```python
Combined_Data.drop_duplicates(subset=['UserId', 'DateTime', 'Event', 'VideoId', 'Date', 'Minute_Of_Day', 'Second', 'State', 'Genre', 'Category', 'ProgramType', 'Country', 'Device', 'OS','Status'],keep='first',inplace=True)
```

```python
Combined_Data['PreviousDateTime'] = Combined_Data['DateTime'].copy()
Combined_Data.PreviousDateTime = Combined_Data.groupby(['UserId'])['PreviousDateTime'].shift(1)
```

```python
Combined_Data['PreviousEvent'] = Combined_Data['Event'].copy()
Combined_Data.PreviousEvent = Combined_Data.groupby(['UserId'])['Event'].shift(1)
```

```python
Combined_Data['SameEventLastDateTime'] = Combined_Data['DateTime'].copy()
Combined_Data.SameEventLastDateTime = Combined_Data.groupby(['UserId','Event'])['PreviousDateTime'].shift(1)
```
