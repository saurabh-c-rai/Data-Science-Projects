```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, accuracy_score, precision_score
from xgboost import XGBRegressor
```

```python
path = '/Users/raisaurabh04/OneDrive/GreyAtom/Practice Dataset/bank_deposit_training.csv'
```

```python
df = pd.read_csv(path, )

df.head()

df.columns = df.columns.str.replace(' ', '_').str.lower()

df.head()

df.isna().sum()

df.fillna(np.nan, inplace=True)
df.isna().sum()
```

```python
df.set_index(keys='serial_number',inplace=True,drop=True)
```

```python
# convert timestamps to datetime to work with time data
df['acquired_date'] = pd.to_datetime(df['acquired_date'])

df['established_date'] = pd.to_datetime(df['established_date'], infer_datetime_format=True)

df.isna().sum()

# Independent variable and dependent variable
X = df.iloc[ : , : -1]
y = df.iloc[ : , -1]

# Train and test data
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.25, random_state=3)
```

```python
def since_date(dataframe):
    time_col = dataframe.select_dtypes(exclude=[np.number, 'O']).columns
    for col_name in time_col:
        new_col_name = 'since_'+ col_name
        dataframe[new_col_name] = (pd.datetime.now() - dataframe[col_name])/np.timedelta64(1, 'Y')
        dataframe.drop(columns=col_name, inplace=True)


since_date(X_train)

since_date(X_val)
```

```python
# Missing values
X_train.fillna(0, inplace=True)
X_val.fillna(0, inplace=True)


# Label encoding
le = LabelEncoder()
cat = X_train.select_dtypes(include='O').columns.tolist()
def transform_categorical(dataframe) :
    for cols in cat:
        dataframe[cols] = le.fit_transform(dataframe[cols])

transform_categorical(X_train)
transform_categorical(X_val)

# One hot encoding
X_train_temp = pd.get_dummies(data=X_train, columns=cat)
X_val_temp = pd.get_dummies(data=X_val, columns=cat)
```

```python
# Shape of train and test data
X_train_temp.shape, X_val_temp.shape

X_train.shape, X_val.shape
```

```python
dt = DecisionTreeRegressor(random_state=5)

dt.fit(X_train, y_train)

accuracy = dt.score(X_val, y_val)

print(accuracy)

y_pred = dt.predict(X_val)

rmse = np.sqrt(mean_squared_error(y_val, y_pred))

print(rmse)
```

```python
xgb = XGBRegressor(max_depth=50, learning_rate=0.83, n_estimators=100)

xgb.fit(X_train, y_train)

accuracy = xgb.score(X_val, y_val)
accuracy
y_pred = xgb.predict(X_val)

rmse = np.sqrt(mean_squared_error(y_val, y_pred))
rmse
```
