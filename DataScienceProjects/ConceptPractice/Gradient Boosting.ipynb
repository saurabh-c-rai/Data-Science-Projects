```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix, classification_report
```

```python
path = '/Users/raisaurabh04/OneDrive/GreyAtom/Practice Dataset/bank_gradient_boosting_pratice.csv'
```

```python
#Loading of the data
data = pd.read_csv(path)

data.head()


#Extracting the features
X = data.iloc[ : , : -1]

#Extracting the target class
y = data['deposit'].copy()

#Splitting the dataset into test and train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

#Fitting of Weak Classifier
dt_clf = DecisionTreeClassifier(max_depth=1, random_state=0)

dt_clf.fit(X_train, y_train)

dt_score = dt_clf.score(X_test, y_test)
print(dt_score)



# Fitting of weak classifier with Adaboost

ada_clf = AdaBoostClassifier(base_estimator=dt_clf, random_state=0)

ada_clf.fit(X_train, y_train)

ada_score = ada_clf.score(X_test, y_test)

print(ada_score)
```

```python
gb_clf = GradientBoostingClassifier(random_state=0)

gb_clf.fit(X_train, y_train)

gb_score = gb_clf.score(X_test, y_test)
print(gb_score)
```

```python
xgb_clf = XGBClassifier(base_estimator=dt_clf,random_state=0)

xgb_clf.fit(X_train, y_train)

gb_score = xgb_clf.score(X_test,y_test)

gb_score
```

```python
path = '/Users/raisaurabh04/OneDrive/GreyAtom/Practice Dataset/customer_assesment_xgboost.csv'
```

```python
df = pd.read_csv(path)

df.head()

X = df.iloc[ : , 1 : -1]

y = df.iloc[ : , -1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_train.columns
```

```python
X_train[X_train['TotalCharges'] == ' ']

X_train['TotalCharges'].replace(' ', np.nan, inplace=True)

X_test['TotalCharges'].replace(' ', np.nan, inplace=True)

X_train['TotalCharges'] = X_train['TotalCharges'].astype('float64')

X_test['TotalCharges'] = X_test['TotalCharges'].astype('float64')

X_train['TotalCharges'].fillna(X_train['TotalCharges'].mean(), inplace=True)

X_test['TotalCharges'].fillna(X_test['TotalCharges'].mean(), inplace=True)

X_train.isna().sum()
X_test.isna().sum()

label_encoder = LabelEncoder()

for cols in X_train.select_dtypes(exclude=np.number):
    X_train[cols] = label_encoder.fit_transform(X_train[cols])
    X_test[cols] = label_encoder.fit_transform(X_test[cols])

mapping_dict = {'Yes' : 1, 'No' : 0}

y_train = y_train.map(mapping_dict)

y_test = y_test.map(mapping_dict)
```

```python
X_train.head()
y_train.head()
X_test.head()
y_test.head()

ada_model = AdaBoostClassifier(random_state=0)

ada_model.fit(X_train, y_train)

y_pred = ada_model.predict(X_test)

ada_cm = confusion_matrix(y_test, y_pred)

ada_cr = classification_report(y_test, y_pred)
ada_score = accuracy_score(y_test, y_pred)
```

```python
ada_cm
ada_cr
ada_score
```

```python
parameters={'learning_rate':[0.1,0.15,0.2,0.25,0.3],
            'max_depth':range(1,3)}
```

```python
xgb_model = XGBClassifier(random_state=0)

xgb_model.fit(X_train, y_train)

y_pred = xgb_model.predict(X_test)

xgb_score = accuracy_score(y_test, y_pred)

xgb_cm = confusion_matrix(y_test, y_pred)

xgb_cr = classification_report(y_test, y_pred)

xgb_score
xgb_cm
xgb_cr
```

```python
clf_model = GridSearchCV(estimator=xgb_model, param_grid=parameters)

clf_model.fit(X_train, y_train)

y_pred = clf_model.predict(X_test)

clf_score = accuracy_score(y_test, y_pred)

clf_cm = confusion_matrix(y_test, y_pred)

clf_cr = classification_report(y_test, y_pred)

clf_score
clf_cm
clf_cr
```
