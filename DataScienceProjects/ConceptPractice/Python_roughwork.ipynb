```python
def golden_ratio(dataframe, col_event, col_total) :
    dataframe['Golden_Ratio'] = dataframe[col_event] / dataframe[col_total]
    max_golden_ratio_row = dataframe.loc[dataframe['Golden_Ratio'].idxmax()]
    max_golden_ratio_country = max_golden_ratio_row['Country_Name']
    max_golden_ratio = max_golden_ratio_row['Golden_Ratio']
    return dataframe, max_golden_ratio, max_golden_ratio_country    
```
